```python
import dash
from dash import Input, Output, State, html, dcc
import datetime
import dash_bootstrap_components as dbc
import pandas as pd
import pathlib
import warnings

warnings.filterwarnings('ignore')

PATH = pathlib.Path(__file__).parent
# app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

layout = dbc.Container([
    html.Br(),
    dbc.Row([
        dbc.Col([
            dcc.Upload(id='upload-image_1',
                       children=html.Div(
                           ['Drag and Drop or ',
                            html.A('Select Files')]),
                       style={
                           'borderStyle': 'dashed',
                           'borderRadius': '5px',
                           'textAlign': 'center',
                           "height": "60px"
                       },
                       multiple=True),
            html.Div(id='output-image-upload_1'), 
            html.Div(id='output_container_1', children=[]),
            html.Br()
        ],
                width=6),
        dbc.Col([
            dcc.Upload(id='upload-image_2',
                       children=html.Div(
                           ['Drag and Drop or ',
                            html.A('Select Files')]),
                       style={
                           'borderStyle': 'dashed',
                           'borderRadius': '5px',
                           'textAlign': 'center',
                           "height": "60px",
                       },
                       multiple=True),
            html.Div(id='output-image-upload_2'), 
            html.Div(id='output_container_2', children=[]),
            html.Br(),
        ],
                width=6),
        dbc.Row([
            dbc.Col([
                html.Label('Living Room : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #        value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Master Bedroom : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Toilet : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Bedroom : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Kitchen : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Dry Balcony : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Dining : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Terrace 1 : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Terrace 2 : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                # dcc.Input(placeholder='',
                #         type = 'text',
                #         value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
            ]),
            dbc.Col([
                html.Label('Living Room : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Master Bedroom : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Toilet : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Bedroom : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Kitchen : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Dry Balcony : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             "width": "18rem",
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Dining : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Terrace 1 : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
                html.Br(),
                html.Br(),
                html.Label('Terrace 2 : Click and Select'),
                html.Br(),
                dbc.Card([dbc.CardHeader('', style={'background': 'white'})],
                         style={
                             'border-color': 'black',
                             'width': '100%',
                             'height': '7vh'
                         }),
                #             dcc.Input(placeholder='',
                #                      type = 'text',
                #                      value= ''),
                dbc.Button("Click",
                           outline=True,
                           color="primary",
                           className="me-1"),
            ]),
        ]),
    ]),
])


def parse_contents(contents, filename, date):
    return html.Div([
        html.H5(filename),
#         html.H6(datetime.datetime.fromtimestamp(date)),
        html.Img(src=contents)
    ])


@app.callback(Output('output-image-upload_1', 'children'),
              Input('upload-image_1', 'contents'),
              State('upload-image_1', 'filename'),
              State('upload-image_1', 'last_modified'))
@app.callback(Output('output-image-upload_2', 'children'),
              Input('upload-image_2', 'contents'),
              State('upload-image_2', 'filename'),
              State('upload-image_2', 'last_modified'))
def update_output(list_of_contents, list_of_names, list_of_dates):
    if list_of_contents is not None:
        children = [
            parse_contents(c, n, d)
            for c, n, d in zip(list_of_contents, list_of_names, list_of_dates)
        ]
        return children
```
